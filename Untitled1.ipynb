```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab-data/train.csv',index_col='PassengerId')
df.head()
```

```python
column_names = df.columns.tolist()
print("Column names:", column_names)
```

```python
# Removing Na values
df['HomePlanet'].fillna('Earth',inplace=True)

x = df.loc[df['VIP']==True]['RoomService'].mean()
df.loc[df['VIP']==True]=df.loc[df['VIP']==True]['RoomService'].fillna(x,inplace=True)

x = df.loc[df['VIP']==False]['RoomService'].mean()
df.loc[df['VIP']==False]=df.loc[df['VIP']==False]['RoomService'].fillna(x,inplace=True)

df['RoomService'].fillna(df['RoomService'].mean(),inplace=True)
print(df['RoomService'].isna().sum())

x = df.loc[df['VIP']==True]['FoodCourt'].mean()
df.loc[df['VIP']==True]=df.loc[df['VIP']==True]['FoodCourt'].fillna(x,inplace=True)

x = df.loc[df['VIP']==False]['FoodCourt'].mean()
df.loc[df['VIP']==False]=df.loc[df['VIP']==False]['FoodCourt'].fillna(x,inplace=True)

df['FoodCourt'].fillna(df['FoodCourt'].mean(),inplace=True)
print(df['FoodCourt'].isna().sum())

x = df.loc[df['VIP']==True]['ShoppingMall'].mean()
df.loc[df['VIP']==True]=df.loc[df['VIP']==True]['ShoppingMall'].fillna(x,inplace=True)

x = df.loc[df['VIP']==False]['ShoppingMall'].mean()
df.loc[df['VIP']==False]=df.loc[df['VIP']==False]['ShoppingMall'].fillna(x,inplace=True)

df['ShoppingMall'].fillna(df['ShoppingMall'].mean(),inplace=True)
print(df["ShoppingMall"].isna().sum())

x = df.loc[df['VIP']==True]['Spa'].mean()
df.loc[df['VIP']==True]=df.loc[df['VIP']==True]['Spa'].fillna(x,inplace=True)

x = df.loc[df['VIP']==False]['Spa'].mean()
df.loc[df['VIP']==False]=df.loc[df['VIP']==False]['Spa'].fillna(x,inplace=True)

df['Spa'].fillna(df['Spa'].mean(),inplace=True)
print(df["Spa"].isna().sum())

x = df.loc[df['VIP']==True]['VRDeck'].mean()
df.loc[df['VIP']==True]=df.loc[df['VIP']==True]['VRDeck'].fillna(x,inplace=True)

x = df.loc[df['VIP']==False]['VRDeck'].mean()
df.loc[df['VIP']==False]=df.loc[df['VIP']==False]['VRDeck'].fillna(x,inplace=True)

df['VRDeck'].fillna(df['VRDeck'].mean(),inplace=True)
print(df["VRDeck"].isna().sum())

```
